```python
import requests

search_url = (
    "https://collectionapi.metmuseum.org/public/collection/v1/search"
    "?hasImages=true&q=painting"
)

r = requests.get(search_url)
r.raise_for_status()

search_data = r.json()

search_data.keys()
```

```python
object_ids = search_data["objectIDs"]
```

```python
object_ids = search_data["objectIDs"][:20]
paintings_with_images = []

# 2. Loop over first 20 IDs
for object_id in object_ids:
    object_url = (
        "https://collectionapi.metmuseum.org/public/collection/v1/objects/"
        f"{object_id}"
    )

    r = requests.get(object_url)
    if r.status_code != 200:
        print(f"Skipped {object_id} (status {r.status_code})")
        continue

    obj = r.json()

    # 3. Only keep paintings with an image URL
    if (
        obj.get("objectName") == "Painting"
        and obj.get("primaryImageSmall")
    ):
        paintings_with_images.append({
            "id": obj["objectID"],
            "title": obj["title"],
            "artist": obj["artistDisplayName"],
            "image_url": obj["primaryImageSmall"],
        })

        print(f"Saved: {obj['title']}")

print("\nTotal saved paintings:", len(paintings_with_images))
```
